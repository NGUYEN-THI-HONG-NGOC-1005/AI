```python
from networkx.algorithms.bipartite.basic import color
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
thoigian=ctrl.Antecedent(np.arange(15,120,5),'thoigian')
com=ctrl.Antecedent(np.arange(150,1300,5),'com')
congsuat=ctrl.Consequent(np.arange(0,1.2,0.5),'congsuat')
thoigian['ngan']=fuzz.trimf(thoigian.universe,[15,15,25])
thoigian['hoi ngan']=fuzz.trimf(thoigian.universe,[15,25,40])
thoigian['vua du']=fuzz.trimf(thoigian.universe,[25,40,80])
thoigian['hoi dai']=fuzz.trimf(thoigian.universe,[40,80,100])
thoigian['dai']=fuzz.trimf(thoigian.universe,[80,100,120])
com['com song']=fuzz.trimf(com.universe,[150,150,300])
com['com nhao']=fuzz.trimf(com.universe,[150,300,450])
com['com chin']=fuzz.trimf(com.universe,[300,450,900])
com['com khet']=fuzz.trimf(com.universe,[450,900,1100])
congsuat['yeu']=fuzz.trimf(congsuat.universe,[0,0,0.3])
congsuat['binh thuong']=fuzz.trimf(congsuat.universe,[0,0.3,0.6])
congsuat['hoi manh']=fuzz.trimf(congsuat.universe,[0.3,0.6,0.9])
congsuat['manh']=fuzz.trimf(congsuat.universe,[0.9,1,1])
thoigian.view()
com.view()
congsuat.view()
congsuat['binh thuong'].view()
rule1=ctrl.Rule(thoigian['ngan'] & com['com song'],congsuat['yeu'])
rule2=ctrl.Rule(thoigian['ngan'] & com['com nhao'],congsuat['yeu'])
rule3=ctrl.Rule(thoigian['ngan'] & com['com khet'],congsuat['manh'])
rule4=ctrl.Rule(thoigian['hoi ngan'] & com['com song'],congsuat['yeu'])
rule5=ctrl.Rule(thoigian['hoi ngan'] & com['com nhao'],congsuat['binh thuong'])
rule6=ctrl.Rule(thoigian['vua du'] & com['com khet'],congsuat['manh'])
rule7=ctrl.Rule(thoigian['vua du'] & com['com song'],congsuat['binh thuong'])
rule8=ctrl.Rule(thoigian['hoi dai'] & com['com nhao'],congsuat['binh thuong'])
rule9=ctrl.Rule(thoigian['hoi dai'] & com['com khet'],congsuat['manh'])
rule10=ctrl.Rule(thoigian['dai'] & com['com chin'],congsuat['hoi manh'])
rule11=ctrl.Rule(thoigian['dai'] & com['com nhao'],congsuat['binh thuong'])
rule12=ctrl.Rule(thoigian['dai'] & com['com song'],congsuat['binh thuong'])
rule13=ctrl.Rule(thoigian['dai'] & com['com khet'],congsuat['manh'])
congsuatvao_ctrl=ctrl.ControlSystem([rule1,rule2,rule4,rule5,rule6,rule7,rule8,rule9,rule10,rule11,rule12,rule13])
congsuatvao=ctrl.ControlSystemSimulation(congsuatvao_ctrl)
congsuatvao.input['thoigian']=47
congsuatvao.input['com']=789
congsuatvao.compute()
print(congsuatvao.output['congsuat'])
congsuat.view(sim=congsuatvao)
```
